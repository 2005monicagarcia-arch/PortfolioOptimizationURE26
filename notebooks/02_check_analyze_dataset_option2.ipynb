```python
import yfinance as yf
import pandas as pd

# Time period: 20 years
start_date = "2006-01-01"
end_date = "2026-06-30"

# Original tickers without VOO
base_tickers = ["AAPL", "MSFT", "NVDA", "SPY", "QQQ", "TLT", "IEF"]

# Option 1: Global Hard-Asset Blend
# Add international stocks + real estate
option1_tickers = base_tickers + ["EFA", "VNQ"]

# Option 2: Core Domestic Value Tilt
# Add financials + energy
option2_tickers = base_tickers + ["XLF", "XLE"]


# Function to download dataset
def download_dataset(tickers, file_name):
    data = yf.download(
        tickers,
        start=start_date,
        end=end_date,
        auto_adjust=False
    )

    # Save full dataset
    data.to_csv(file_name)

    print(f"\nDownloaded: {file_name}")
    print(f"Tickers: {tickers}")
    print(f"Shape: {data.shape}")

    # Check missing values for Adjusted Close
    adj_close = data["Adj Close"]
    print("\nMissing values in Adjusted Close:")
    print(adj_close.isna().sum())

    return data


# Download Option 1
data_option1 = download_dataset(
    option1_tickers,
    "portfolio_option1_EFA_VNQ_20years.csv"
)

# Download Option 2
data_option2 = download_dataset(
    option2_tickers,
    "portfolio_option2_XLF_XLE_20years.csv"
)
```

```python
import pandas as pd
df2 = pd.read_csv("portfolio_option2_XLF_XLE_20years.csv", header=[0, 1], index_col=0, parse_dates=True)
df2.head()
```

```python
adj_close2 = df2["Adj Close"]
print("Missing values in Adjusted Close:")
print(adj_close2.isna().sum())
```

```python
daily_returns2 = adj_close2.pct_change().dropna()

summary2 = pd.DataFrame({
    "Annual Return %": daily_returns2.mean() * 252 * 100,
    "Annual Volatility %": daily_returns2.std() * (252 ** 0.5) * 100,
    "Min Daily Return": daily_returns2.min() * 100,
    "Max Daily Return": daily_returns2.max() * 100,
})
summary2.round(2)
```

```python
correlation_matrix2 = daily_returns2.corr()
correlation_matrix2.round(2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.imshow(correlation_matrix2, aspect="auto")
plt.colorbar()
plt.xticks(range(len(correlation_matrix2.columns)), correlation_matrix2.columns, rotation=45)
plt.yticks(range(len(correlation_matrix2.columns)), correlation_matrix2.columns)
plt.title("Correlation Matrix of Daily Returns — Option 2 (XLF/XLE)")
plt.savefig("correlation_heatmap_option2.png", dpi=150, bbox_inches="tight")
plt.show()
```
