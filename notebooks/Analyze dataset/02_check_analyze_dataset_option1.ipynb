```python
import pandas as pd

df = pd.read_csv(
    "portfolio_option1_EFA_VNQ_20years.csv",
    header=[0, 1],
    index_col=0,
    parse_dates=True
)

df.head()
```

```python
df.columns
```

```python
adj_close = df["Adj Close"]

adj_close.head()
```

```python
list(adj_close.columns)
```

```python
print("Start date:", adj_close.index.min())
print("End date:", adj_close.index.max())
```

```python
print("Shape:", adj_close.shape)
```

```python
adj_close.isna().sum()
```

```python


import matplotlib.pyplot as plt

adj_close.plot(figsize=(12, 6))
plt.title("Adjusted Close Prices")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
normalized_prices = adj_close / adj_close.iloc[0] * 100

normalized_prices.plot(figsize=(12, 6))
plt.title("Normalized Adjusted Close Prices")
plt.xlabel("Date")
plt.ylabel("Growth of $100")
plt.show()
```

```python
daily_returns = adj_close.pct_change().dropna()

daily_returns.head()
```

```python
annual_return = daily_returns.mean() * 252
annual_volatility = daily_returns.std() * (252 ** 0.5)
min_daily_return = daily_returns.min()
max_daily_return = daily_returns.max()

summary = pd.DataFrame({
    "Annual Return": annual_return,
    "Annual Volatility": annual_volatility,
    "Min Daily Return": min_daily_return,
    "Max Daily Return": max_daily_return
})

summary
```

```python
summary_percent = summary * 100
summary_percent.round(2)
```

```python
summary_percent.sort_values("Annual Volatility", ascending=False).round(2)
```

```python
correlation_matrix = daily_returns.corr()

correlation_matrix.round(2)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.imshow(correlation_matrix, aspect="auto")
plt.colorbar()

plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)

plt.title("Correlation Matrix of Daily Returns")
plt.show()
```
